# **1. Introduction**



### Dataset Overview
- **Content**: The dataset consists of financial data representing grants given to different countries and regions, excluding technical cooperation funds. These grants are presumably from international organizations, governments, or other financial entities that provide aid or assistance.

- **Scope**: The data spans several decades, providing historical insight into the flow of financial aid across the globe. The time frame typically ranges from the 1960s up to the recent past, allowing for longitudinal analyses of trends and patterns.

- **Granularity**: Each record in the dataset pertains to a specific country or region and includes yearly data points that represent the amount of grants received in current US dollars for that year.

- **Key Attributes**:
  - **Country Name**: The name of the country or region receiving the grant.
  - **Country Code**: A unique identifier for the country or region.
  - **Indicator Name**: Describes the type of financial data, in this case, grants excluding technical cooperation.
  - **Indicator Code**: A unique code for the financial indicator.
  - **Year Columns (1960, 1961, ..., 2021)**: Each column represents a specific year and contains the grant amount received by the country or region in that year.

- **Purpose**: This dataset can be used to analyze financial aid distribution patterns, assess the impact of global events on aid, and understand the economic relationships between donor entities and recipient countries.

- **Applications**:
  - **Economic Research**: To study the trends in international aid and its correlation with global economic changes or events.
  - **Policy Analysis**: To inform policy-making by identifying which countries have been the major beneficiaries of aid and which have received less, potentially indicating their level of need or geopolitical significance.
  - **Humanitarian Efforts**: To evaluate the distribution of aid in response to crises, conflicts, or natural disasters.

- **Analytical Potential**:
  - **Trend Analysis**: Observing how grant amounts change over time.
  - **Comparative Analysis**: Comparing grant receipts across different countries or regions.
  - **Correlation Study**: Investigating the relationship between grant distribution and other economic indicators or events.

The data holds significant value for those interested in international development, economics, and global finance, providing a quantitative basis for understanding and improving global aid strategies.

# **Inference**

***Model Selection***

The predictive model chosen for forecasting annual grants received by Afghanistan is based on the Linear Regression algorithm. Linear Regression was selected due to its robustness, simplicity, and the interpretability of its results, which are essential features for a preliminary study in economic forecasting. This statistical method assumes a linear relationship between the independent variables (grant amounts from the previous five years) and the dependent variable (grant amount for the subsequent year).

Linear Regression is particularly suited for this analysis due to the continuous nature of the target variable and its capacity to provide a clear quantifiable relationship between past and future grants. The model's efficacy in handling time series data was enhanced by structuring the dataset into a time window format, allowing the model to consider temporal patterns and trends in grant distribution.

```python
X, y = inference.prepare_data('Afghanistan', 5)
```

***Training and Testing***

The data was split into training and testing sets, with the final year's data reserved for testing to evaluate the model's predictive performance. This approach allows for an assessment of the model's generalization capabilities on unseen data.

```python
# Train the model
X_train, X_test = X[:-1], X[-1:]
y_train, y_test = y[:-1], y[-1:]
```

***Model Training***

A regression model was trained on the historical data, using grant amounts from the previous years as input features. The model was tasked with learning the relationship between these features and the target variable, which is the grant amount for the next year.

```python
model = inference.train_model(X_train, y_train)
```

***Prediction and Evaluation***

Post-training, the model was used to predict the grant amount for the test year. The forecasted value was then compared against the actual grant received using the Root Mean Squared Error (RMSE) metric. RMSE provides a measure of the average magnitude of the prediction errors, thus offering a quantitative evaluation of the model's accuracy.

```python
# Make predictions
y_pred = inference.make_predictions(model, X_test.reshape(1, -1))
```

```python
# Evaluate the model
rmse = inference.evaluate_model(y_test, y_pred)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

***Results***

The results, visualized through a scatter plot, indicate that the model's prediction was in close agreement with the actual grant amount for the test year. This preliminary outcome is promising and suggests that the model could serve as a viable tool for forecasting financial aid requirements.

```python
# Plot the predictions against actual values
inference.plot_predictions([y_test[0]], [y_pred[0]])
```

**Prediction Close to Actual:** The predicted grant amount is quite close to the actual amount, as indicated by the proximity of the orange "X" to the blue dot. This suggests that the model has a relatively good performance in forecasting the grant amount for the given year.

**Single Data Point:** Since there is only one data point for both actual and predicted values, it's likely that this prediction is for a specific year (likely the latest year in the dataset) using historical data from the preceding years.

**Scale of Values:** Both the actual and predicted values are in the range of billions of US dollars (as indicated by the y-axis label), which underlines the substantial amount of financial aid involved.

**Model Performance:** While the plot indicates a single prediction, the closeness to the actual value suggests that the model could be useful for forecasting purposes. However, it's important to consider the performance across multiple years or using cross-validation to get a more robust evaluation of the model's accuracy.

## **Evaluating the Research Question with the Model Output**

***Effectiveness of Prediction:***

The linear regression model built on historical grant data for Afghanistan did provide a prediction for future grants (for the year 2021 in our case).
However, the RMSE value indicated that while the model's prediction was relatively close to the actual grants received, there was a notable discrepancy. This suggests that while historical data can be indicative, it might not always be completely accurate in forecasting future financial aid requirements.

***Insights for Strategic Planning:***

The model's output can offer preliminary insights into trends and potential future needs. For instance, if the model consistently predicts increasing or decreasing trends, this might signal changing financial aid requirements.
However, the limitations of the model should be acknowledged. Factors like political changes, economic shifts, global events (like pandemics or economic crises), and changes in international policies can significantly influence grant allocations and are not always captured in historical data.
